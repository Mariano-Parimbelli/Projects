## evaluation metric

```python
import numpy as np

def apk(actual, predicted, k=10):
 
    if len(predicted)>k:
        predicted = predicted[:k]

    score = 0.0
    num_hits = 0.0

    for i,p in enumerate(predicted):
        if p in actual and p not in predicted[:i]:
            num_hits += 1.0
            score += num_hits / (i+1.0)

    if not actual:
        return 0.0

    return score / min(len(actual), k)

def mapk(actual, predicted, k=10):
 
    return np.mean([apk(a,p,k) for a,p in zip(actual, predicted)])
```

```python
from sklearn.ensemble import BaggingClassifier
# Define la función de evaluación
def mapk_eval(y_true, y_pred):
    k = 5
    y_pred_top_k = np.argsort(y_pred, axis=1)[:, -k:][:, ::-1]
    return mapk(y_true.apply(lambda x: [x]), y_pred_top_k)



# Define los clasificadores y sus espacios de búsqueda de hiperparámetros
classifiers = {
    'Extra Trees': (ExtraTreesClassifier(random_state=42), {
        'n_estimators': [10,40,100,300, 500],
        'max_features': ['sqrt', 'log2', None]
    }),
    'Random Forest': (RandomForestClassifier(random_state=42), {
        'n_estimators': [10,40,100,300, 500],
        'max_features': ['sqrt', 'log2', None]
    }),
    'AdaBoost': (AdaBoostClassifier(random_state=42), {
        'n_estimators': [10,40,100,300, 500],
        'learning_rate': [0.01, 0.1, 1.0]
    }),

    'Logistic Regression': (LogisticRegression(random_state=42, max_iter=9000), {
        'penalty': ['l2'],
        'C': [0.1, 1.0, 10.0]
    }),
    'K-NN': (KNeighborsClassifier(), {
        'n_neighbors': [3, 5, 7],
        'weights': ['uniform', 'distance']
    }),
    'Decision Tree': (DecisionTreeClassifier(random_state=42), {
        'max_depth': [5, 10, 20, None],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4]
    }),
    'Gradient Boosting': (GradientBoostingClassifier(random_state=42), {
        'n_estimators': [10, 40, 100, 300, 500],
        'learning_rate': [0.01, 0.1, 1.0],
        'max_depth': [3, 5, 7],
        'max_features': ['sqrt', 'log2', None]
    }),
       
    'LightGBM': (LGBMClassifier(random_state=42), {
        'n_estimators': [10, 40, 100, 300, 500],
        'learning_rate': [0.01, 0.1, 1.0],
        'num_leaves': [31, 127, 255, 511]
    }),
    'XGBoost': (XGBClassifier(random_state=42), {
        'n_estimators': [10, 40, 100, 300, 500],
        'learning_rate': [0.01, 0.1, 1.0],
        'max_depth': [3, 5, 7, 9]
    }),
    'CatBoost': (CatBoostClassifier(random_state=42, silent=True), {
        'n_estimators': [10, 40, 100, 300, 500],
        'learning_rate': [0.01, 0.1, 1.0],
        'depth': [3, 5, 7, 9]
    })
}

#sca los mejores hiperparámetros para cada clasificador
for name, (clf, params) in classifiers.items():
    scorer = make_scorer(mapk_eval, needs_proba=True)
    search = GridSearchCV(clf, params, cv=5, scoring=scorer)
    search.fit(X_val, y_val)
    print(f"{name}: Best parameters: {search.best_params_}, Best score: {search.best_score_}")
    print("_____________________________________________________")
```

```python
from catboost import CatBoostClassifier
from sklearn.model_selection import GridSearchCV
import numpy as np

# Definir los hiperparámetros a probar
param_grid = {
    'learning_rate': [0.1, 0.5],
    'depth': [3, 5, 7],
    'n_estimators': [10,50, 100, 200,250,300,350,400,500],
    'depth': [3, 5,7,9]

}

# Crear el modelo de CatBoost
model = CatBoostClassifier(random_state=42, silent=True)

# Realizar la búsqueda de hiperparámetros con validación cruzada
grid_search = GridSearchCV(model, param_grid, cv=5, scoring='neg_log_loss', n_jobs=-1)
grid_search.fit(X_val_train, y_val_train)

# Seleccionar el modelo con los mejores hiperparámetros
best_model = grid_search.best_estimator_

# Realizar predicciones en el conjunto de prueba
y_pred = best_model.predict_proba(X_val_test)

# Obtener las k mejores predicciones para cada consulta
k = 10
y_pred_sorted = np.argsort(y_pred, axis=1)[:, -k:][:, ::-1]

# Calcular la métrica MAP@k en el conjunto de prueba
mapk_score = mapk(y_val_test.apply(lambda x: [x]), y_pred_sorted)

print(f"El mejor modelo tiene los siguientes hiperparámetros: {best_model.get_params()}")
print(f"El score MAP@{k} del modelo en el conjunto de prueba es: {mapk_score}")
```
